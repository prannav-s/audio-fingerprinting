```python
from pydub import AudioSegment
import os

def convert_mp3_to_wav(input_folder, output_folder):
    """
    Converts all MP3 files in the input folder to WAV format and saves them in the output folder.
    """
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)
    
    for file_name in os.listdir(input_folder):
        if file_name.endswith(".mp3"):
            try:
                mp3_path = os.path.join(input_folder, file_name)
                wav_path = os.path.join(output_folder, file_name.replace(".mp3", ".wav"))
                
                # Convert MP3 to WAV
                audio = AudioSegment.from_mp3(mp3_path)
                audio.export(wav_path, format="wav")
                
                print(f"Converted {file_name} to {wav_path}")
            except Exception as e:
                print(f"Error converting {file_name}: {e}")

# Example usage
convert_mp3_to_wav("input_mp3_folder", "output_wav_folder")
```

```python
from pydub import AudioSegment
from pydub.utils import make_chunks
import os

def split_audio_by_class(input_folder, output_folder, chunk_length_ms=3000):
    """
    Splits audio files into smaller chunks and organizes them into class-based directories.
    """
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)
    
    for file_name in os.listdir(input_folder):
        if file_name.endswith(".wav"):
            try:
                # Create a class-specific subdirectory
                class_folder = os.path.join(output_folder, file_name[:-4])  # Use the file name as the class
                if not os.path.exists(class_folder):
                    os.makedirs(class_folder)
                
                # Process the file
                file_path = os.path.join(input_folder, file_name)
                audio = AudioSegment.from_file(file_path)
                chunks = make_chunks(audio, chunk_length_ms)

                # Save chunks in the class directory
                for i, chunk in enumerate(chunks):
                    chunk_name = f"{file_name[:-4]}_chunk{i}.wav"
                    chunk.export(os.path.join(class_folder, chunk_name), format="wav")
                
                print(f"Processed {file_name}, created {len(chunks)} chunks.")
            
            except Exception as e:
                print(f"Error processing {file_name}: {e}")
        else:
            print(f"Skipped non-WAV file: {file_name}")

# Example usage
split_audio_by_class("output_wav_folder", "output_chunks_folder", chunk_length_ms=3000)
```

```python
import os
import tensorflow as tf

def load_snippets_from_directory(base_dir):
    """
    Load audio chunks from directories and create a dataset with labels.
    """
    file_paths = []
    labels = []

    for label, class_name in enumerate(os.listdir(base_dir)):
        class_dir = os.path.join(base_dir, class_name)
        if os.path.isdir(class_dir):
            for file_name in os.listdir(class_dir):
                if file_name.endswith(".wav"):
                    file_paths.append(os.path.join(class_dir, file_name))
                    labels.append(label)  # Numeric label for the class

    return file_paths, labels

# Example usage
base_dir = "output_chunks_folder"
file_paths, labels = load_snippets_from_directory(base_dir)
print("File Paths:", file_paths[:5])
print("Labels:", labels[:5])
```

```python
from sklearn.model_selection import train_test_split

train_file_paths, val_file_paths, train_labels, val_labels = train_test_split(
    file_paths, labels, test_size=0.2, random_state=42
)
```

```python
def prepare_dataset(file_paths, labels, target_height=4125, target_width=257):
    """
    Prepares the dataset by loading audio files, generating spectrograms,
    and resizing them to consistent dimensions.
    """
    dataset = tf.data.Dataset.from_tensor_slices((file_paths, labels))

    def preprocess(file_path, label):
        """
        Preprocess a single audio file:
        - Load the WAV file
        - Compute the spectrogram
        - Resize the spectrogram to target dimensions
        """
        audio = tf.io.read_file(file_path)
        audio, _ = tf.audio.decode_wav(audio, desired_channels=1)
        audio = tf.squeeze(audio, axis=-1)  # Remove unnecessary dimensions
        spectrogram = tf.signal.stft(audio, frame_length=320, frame_step=32)
        spectrogram = tf.abs(spectrogram)
        
        # Ensure the spectrogram has 3 dimensions (height, width, channels)
        spectrogram = tf.expand_dims(spectrogram, axis=-1)  # Add a channel dimension
        
        # Resize the spectrogram to fixed dimensions
        spectrogram = tf.image.resize(spectrogram, [target_height, target_width])
        
        return spectrogram, label

    dataset = dataset.map(preprocess)
    dataset = dataset.batch(32).prefetch(tf.data.AUTOTUNE)
    return dataset

# Example usage
train_dataset = prepare_dataset(train_file_paths, train_labels, target_height=4125, target_width=257)
val_dataset = prepare_dataset(val_file_paths, val_labels, target_height=4125, target_width=257)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, Dense, Flatten, Dropout, MaxPooling2D
from tensorflow.keras.layers import GlobalAveragePooling2D
def create_model(input_shape, num_classes):
    model = Sequential([
        Conv2D(32, (3, 3), activation='relu', input_shape=input_shape),
        MaxPooling2D((2, 2)),
        Dropout(0.3),
        Conv2D(64, (3, 3), activation='relu'),
        MaxPooling2D((2, 2)),
        Dropout(0.3),
        GlobalAveragePooling2D(),  # Handles variable input sizes
        Dense(128, activation='relu'),
        Dropout(0.3),
        Dense(num_classes, activation='softmax')
    ])
    model.compile(optimizer='adam',
                  loss='sparse_categorical_crossentropy',
                  metrics=['accuracy'])
    return model

input_shape = (1491, 257, 1)  # Example dimensions
num_classes = len(set(labels))
model = create_model(input_shape, num_classes)
model.summary()
```

```python
import numpy as np
import tensorflow as tf

def predict_song(audio_path, model, target_height=4125, target_width=257, chunk_duration=3, sample_rate=16000):
    """
    Predict the class of a full-length song using a sliding window approach.
    """
    # Load the full audio file
    audio = tf.io.read_file(audio_path)
    audio, _ = tf.audio.decode_wav(audio, desired_channels=1)
    audio = tf.squeeze(audio, axis=-1)

    # Calculate chunk size in samples
    chunk_size = int(chunk_duration * sample_rate)

    # Generate predictions for each chunk
    predictions = []
    for start in range(0, len(audio) - chunk_size + 1, chunk_size):
        # Extract a chunk of the audio
        chunk = audio[start:start + chunk_size]

        # Compute the spectrogram
        spectrogram = tf.signal.stft(chunk, frame_length=320, frame_step=32)
        spectrogram = tf.abs(spectrogram)

        # Resize the spectrogram to match the model input dimensions
        spectrogram = tf.expand_dims(spectrogram, axis=-1)
        spectrogram = tf.image.resize(spectrogram, [target_height, target_width])
        spectrogram = tf.expand_dims(spectrogram, axis=0)  # Add batch dimension

        # Predict the class for this chunk
        prediction = model.predict(spectrogram, verbose=0)
        predictions.append(prediction)

    # Aggregate predictions
    predictions = np.array(predictions)
    aggregated_prediction = np.argmax(np.mean(predictions, axis=0))

    return aggregated_prediction
```

```python
from flask import Flask, render_template, request
import os
from tensorflow.keras.models import load_model
import tensorflow as tf
import numpy as np

# Initialize Flask app
app = Flask(__name__)
UPLOAD_FOLDER = "uploads"
os.makedirs(UPLOAD_FOLDER, exist_ok=True)

# Load the trained model
model = load_model("song_classification_model.h5")

# Define class-to-song mapping
class_to_song = {
    0: "Song Title 1",
    1: "Song Title 2",
    2: "Song Title 3",
}

# Define predict_song function
def predict_song(audio_path, loaded_model, target_height=4125, target_width=257, chunk_duration=3, sample_rate=16000):
    """
    Predict the class of a full-length song using a sliding window approach.
    """
    audio = tf.io.read_file(audio_path)
    audio, _ = tf.audio.decode_wav(audio, desired_channels=1)
    audio = tf.squeeze(audio, axis=-1)
    chunk_size = int(chunk_duration * sample_rate)
    predictions = []
    for start in range(0, len(audio) - chunk_size + 1, chunk_size):
        chunk = audio[start:start + chunk_size]
        spectrogram = tf.signal.stft(chunk, frame_length=320, frame_step=32)
        spectrogram = tf.abs(spectrogram)
        spectrogram = tf.expand_dims(spectrogram, axis=-1)
        spectrogram = tf.image.resize(spectrogram, [target_height, target_width])
        spectrogram = tf.expand_dims(spectrogram, axis=0)
        prediction = loaded_model.predict(spectrogram, verbose=0)
        predictions.append(prediction)
    predictions = np.array(predictions)
    aggregated_prediction = np.argmax(np.mean(predictions, axis=0))
    return aggregated_prediction

# Define routes
@app.route("/")
def index():
    return render_template("index.html")

@app.route("/predict", methods=["POST"])
def predict():
    if "file" not in request.files:
        return "No file uploaded", 400
    file = request.files["file"]
    if file.filename == "":
        return "No file selected", 400
    file_path = os.path.join(UPLOAD_FOLDER, file.filename)
    file.save(file_path)

    # Predict the song
    predicted_class = predict_song(file_path, model)
    predicted_song = class_to_song.get(predicted_class, "Unknown Song")
    return render_template("result.html", song=predicted_song)

if __name__ == "__main__":
    app.run(debug=True)
```
